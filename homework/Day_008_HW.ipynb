```python
import os
import numpy as np
import pandas as pd

dir_data = './data/'
```

```python
f_app_train = os.path.join(dir_data,'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
fig = plt.figure(figsize=(18,6)) #外面
miss_train = pd.DataFrame((app_train.isnull().sum())*100/app_train.shape[0]).reset_index()
miss_train['type'] = 'train' #標記所有column的資料來自train
missing = pd.concat([miss_train],axis=0) #不懂???
ax = sns.pointplot('index',0,data=missing,hue='type') #裡面
plt.xticks(rotation=90,fontsize=7) #x軸文字轉向以及大小
plt.title('Percentage of Missing values in application train')
plt.ylabel('PERCENTAGE')
plt.xlabel('COLUMN')
ax.set_facecolor('k') #裡面的背景（黑）
fig.set_facecolor('lightgrey') #外面的背景（灰）
```

```python
dtypes_df = app_train.dtypes.reset_index()
dtypes_df.columns = ['Count', 'Column Type']
dtypes_df = dtypes_df.groupby('Column Type').aggregate('count').reset_index()
dtypes_df
# app_train.dtypes.value_counts()
```

```python
# object 類別變項中各有幾類
app_train.select_dtypes('object').apply(pd.Series.nunique,axis=0)
```

```python
print('mean:',app_train['TARGET'].mean())
print('sd:',app_train['TARGET'].std())
plot1 = plt.figure(figsize=(3,3))
app_train['TARGET'].value_counts().plot(kind='bar')
plt.title('target')
plt.xlabel('afford')
plt.ylabel('count')
plot1.set_facecolor('lightgrey')
```

```python
plot2 = plt.figure(figsize=(3,3))
app_train['CODE_GENDER'].value_counts().plot(kind='pie')
plt.title('gender')
plt.ylabel('count')
plot2.set_facecolor('lightgrey')
```

```python
plot3 = plt.figure(figsize=(3,3))
pd.value_counts(app_train['NAME_CONTRACT_TYPE'].values.flatten()).plot(kind='bar')
plt.title('contract type')
plot3.set_facecolor('lightgray')
```

```python
plot3 = plt.figure(figsize=(5,3))
(pd.value_counts(app_train['NAME_EDUCATION_TYPE'].values.flatten())*100/app_train.shape[0]).plot(kind='bar')
plt.title('education')
plt.xlabel('type')
plt.ylabel('percent')
plot3.set_facecolor('lightgray')
```

```python
plot3 = plt.figure(figsize=(5,3))
app_train['NAME_HOUSING_TYPE'].value_counts().plot(kind='bar')
plt.title('house type')
plot3.set_facecolor('lightgray')
```
